```
import populartimes as maps
my_key = "your_key" # insert an key to access google maps api
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
```

```
path = "/content/drive/My Drive/carreira/hackaton/data/supermercados - Página1.csv"
df = pd.read_csv(path)
```

```
col = ['name', 'coordinates', 'address']
week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
for i in week:
  for j in range(24):
    col.append("pop_time_"+ i + "_" + str(j))

print(col)
```

```
local = []
for i in range(df.id.size):
  my_loc = df.id[i]
  local.append(maps.get_id(my_key, my_loc))
  print(local[i])
```
